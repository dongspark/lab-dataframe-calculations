#### Cleanup `Name` that contain "Mega".

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names. For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

```python
pokemon['Name'].value_counts()
```

```python
def clean_mega(x):
    if ('Meganium' in x):
        return x
    elif ('Mega' in x ):
        lastname=  x.split(" ")[1]
        return ('Mage' + ' '+ lastname)
    else:
        return x
    
pokemon['Name'] = pokemon['Name'].apply(clean_mega)  
```

#### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Rules:

* If both `Type 1` and `Type 2` have valid values, the `Combo Type` value should contain both values in the form of `<Type 1> <Type 2>`. For example, if `Type 1` value is `Grass` and `Type 2` value is `Poison`, `Combo Type` will be `Grass-Poison`.

* If `Type 1` has valid value but `Type 2` is not, `Combo Type` will be the same as `Type 1`. For example, if `Type 1` is `Fire` whereas `Type 2` is `NaN`, `Combo Type` will be `Fire`.

```python
pokemon.isna().sum()
```

```python
pokemon.fillna('',inplace=True)#since there are some nan in type2,we can just fill it with white space
```

```python
pokemon.isna().sum()
```

```python
# your code here
pokemon['Combo Typo'] = pokemon.agg(lambda x: f"{x['Type 1']}  {x['Type 2']}",axis=1)
```

```python
pokemon
```

#### Identify the pokemon whose `A/D Ratio` are among the top 5.

```python
# your code here
#display(pokemon.groupby(['Name']).agg({'A/D Ratio':sum}))
top5=pokemon.sort_values(['A/D Ratio','Name'],ascending=False).groupby('Name').apply(lambda x:x).reset_index()
top5.head(5)#don't know why but after I add apply(lambda x:x),finally it can be be re_indexed.
```

```python
one_hot = pokemon[['Combo Typo']]
unique_values =['Grass','Poison','Fire','Flying','Dragon','Water','Bug','Normal','Electric','Ground','Fairy','Fighting','Psychic','Rock','Steel','Ice','Ghost','Dark','']
from sklearn.preprocessing import OneHotEncoder
#encoder = OneHotEncoder(drop='first').fit(one_hot)
encoder = OneHotEncoder(categories=[unique_values]*one_hot.shape[1]).fit(one_hot)
cols = encoder.get_feature_names(input_features = one_hot.columns)
col_encoded = pd.DataFrame(encoder.transform(one_hot).toarray(),columns=cols)
col_encoded
```
